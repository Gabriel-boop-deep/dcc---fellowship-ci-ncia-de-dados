```python
import pandas as pd
import plotly.express as px
```

```python
projecao_pop = pd.read_excel('C:\dcc-fellowship-ciencia-de-dados\projecao_populacao.xlsx')
```

```python
projecao_pop = projecao_pop[projecao_pop['Região'] == 'Brasil']
```

```python
consumo_brasil = pd.read_excel('C:\dcc-fellowship-ciencia-de-dados\pof_rendimento_regiao_2_copia.xlsx')
```

```python
consumo_brasil = consumo_brasil[consumo_brasil['Região'] == 'Brasil'][['Items', 'Região', 'Consumo médio']].reset_index(drop = True)
consumo_brasil.rename(columns = {'Consumo médio':'Consumo médio pcpt (kg)'})
```

```python
projecao_consumo_brasil = pd.merge(projecao_pop, consumo_brasil, how = 'inner', on = 'Região')
```

```python
projecao_consumo_brasil.head()
```

```python
projecao_consumo_brasil['Demanda pop (ton)'] = (projecao_consumo_brasil['Populacao']*projecao_consumo_brasil['Consumo médio'])/1000
```
